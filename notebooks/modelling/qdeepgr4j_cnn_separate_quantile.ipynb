```python
def create_sequence(X, y, window_size):

        assert window_size is not None, "Window size cannot be NoneType."

        # Create empyty sequences
        Xs, ys = [], []

        # Add sequences to Xs and ys
        for i in range(1, len(X) - window_size):
            Xs.append(torch.concat([
                                    X[i: (i + window_size)], 
                                    y[i-1: (i + window_size - 1)]
                                ], dim=1)
                    )
            ys.append(y[i + window_size - 1])

        Xs, ys = torch.stack(Xs), torch.stack(ys)
        Xs = torch.unsqueeze(Xs, dim=1)

        return Xs, ys
```

```python
def val_step(model, dl, q):
    total_loss = 0.
    model.eval()
    for j, (X, y) in enumerate(dl, start=1):
        X, y = X.to(device), y.to(device)
        y_hat = model(X)
        batch_loss = tilted_loss(q, y[:, 0], y_hat[:, 0])
        total_loss += batch_loss
    return (total_loss/j).detach()


def train_step(model, dl, q, opt):
    total_loss = 0.
    model.train()
    for i, (X, y) in enumerate(dl, start=1):
        opt.zero_grad()
        X, y = X.to(device), y.to(device)
        y_hat = model(X)
        batch_loss = tilted_loss(q, y[:, 0], y_hat[:, 0])
        total_loss += batch_loss
        batch_loss.backward()
        opt.step()
    return (total_loss/i).detach()


# def evaluate_preds(model, ds, batch_size, x_scaler=None, y_scaler=None, quantiles=[0.5]):
#     # Evaluate on train data
#     model.eval()
#     dl = torchdata.DataLoader(ds, 
#                               batch_size=batch_size,
#                               shuffle=False)

#     # Empty list to store batch-wise tensors
#     P = []
#     ET = []
#     Q = []
#     Q_hat = []

#     for i, (X, y) in enumerate(dl, start=1):
        
#         X = X.to(device)
#         y_hat = model(X).cpu()
#         X = X.cpu()

#         Q.append((y*y_sigma+y_mu).detach().numpy())
#         Q_hat.append((y_hat*y_sigma+y_mu).detach().numpy())
        
#         X_inv = X[:, 0, -1, :-1]*prod_store.sigma+prod_store.mu
        
#         P.append((X_inv[:, 0]).detach().numpy())
#         ET.append((X_inv[:, 1]).detach().numpy())
    
#     P = np.concatenate(P, axis=0)
#     ET = np.concatenate(ET, axis=0)
#     Q = np.concatenate(Q, axis=0).flatten()
#     Q_hat = np.clip(np.concatenate(Q_hat, axis=0), 0, None)

#     return evaluate(P, ET, Q, Q_hat, quantiles=quantiles)
```

```python
in_channels = 1
n_features = 10
n_filters = (16, 16, 8)
dropout = 0.1
quantiles = [0.05, 0.50, 0.95]
n_quantiles = len(quantiles)
```

```python
n_epoch = 500
lr = 5e-4
```

```python
models = [ConvNet(n_ts=window_size, 
                n_features=n_features,
                n_channels=in_channels,
                out_dim=1,
                n_filters=n_filters,
                dropout_p=dropout).to(device) for i in range(n_quantiles)]
```

```python
def tilted_loss(q, y, f):
    e = (y-f)
    return torch.mean(torch.maximum(q*e, (q-1)*e))
```

```python
def train_model(model, q):
    # Create optimizer and loss instance
    opt = torch.optim.Adam(model.parameters(), 
                           lr=lr, 
                           betas=(0.89, 0.97), 
                           weight_decay=0.02)
    
    # Early stopping
    early_stopper = EarlyStopper(patience=10, min_delta=0.01)
    pbar = tqdm(range(1, n_epoch+1))
    
    train_loss_list = []
    
    for epoch in pbar:
    
        # Train step
        train_loss = train_step(model, train_dl, q, opt)
    
        # Validation step
        val_loss = val_step(model, val_dl, q)
        
        pbar.set_description(f"""Epoch {epoch} loss: {train_loss.cpu().numpy():.4f} val_loss: {val_loss.cpu().numpy():.4f}""")
    
        if early_stopper.early_stop(val_loss):
            break
    
        train_loss_list.append(train_loss.cpu().numpy())

    return np.array(train_loss_list) 
```

```python
for i, q in enumerate(quantiles):
    loss = train_model(models[i], q)
    plt.plot(np.arange(len(loss)), loss, label=f"{q:.2f}")
    plt.legend()
```

```python
def evaluate_preds(models, ds, batch_size, x_scaler=None, y_scaler=None, quantiles=[0.5]):
    
    # Set model state to evaluation
    for model in models:
        model.eval()
        
    # Evaluate on train data
    dl = torchdata.DataLoader(ds, 
                              batch_size=batch_size,
                              shuffle=False)

    # Empty list to store batch-wise tensors
    P = []
    ET = []
    Q = []
    Q_hat = []

    for i, (X, y) in enumerate(dl, start=1):
        
        X = X.to(device)
        y_hat = [model(X).cpu() for model in models]
        y_hat = torch.concat(y_hat, dim=1)
        X = X.cpu()

        Q.append((y*y_sigma+y_mu).detach().numpy())
        Q_hat.append((y_hat*y_sigma+y_mu).detach().numpy())
        
        X_inv = X[:, 0, -1, :-1]*prod_store.sigma+prod_store.mu
        
        P.append((X_inv[:, 0]).detach().numpy())
        ET.append((X_inv[:, 1]).detach().numpy())
    
    P = np.concatenate(P, axis=0)
    ET = np.concatenate(ET, axis=0)
    Q = np.concatenate(Q, axis=0).flatten()
    Q_hat = np.clip(np.concatenate(Q_hat, axis=0), 0, None)

    return evaluate(P, ET, Q, Q_hat, quantiles=quantiles)
```

```python
evaluate_preds(models, val_ds, batch_size=256, quantiles=quantiles)
```
